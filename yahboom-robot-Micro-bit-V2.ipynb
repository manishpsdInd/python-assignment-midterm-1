```python
import uflash
import serial
import time

# Define the script you want to run on the Micro:bit
script = """
from microbit import *

while True:
    display.set_pixel(2, 2, 9)
    sleep(500)
    display.set_pixel(2, 2, 0)
    sleep(500)
"""

# Save the script to a file
script_filename = 'main.py'
with open(script_filename, 'w') as f:
    f.write(script)

# Flash the script to the Micro:bit
uflash.flash(script_filename)

# Replace with your actual port (e.g., 'COM3', '/dev/ttyACM0', '/dev/tty.usbmodemXXXX')
microbit_port = 'COM3'
baud_rate = 115200

# Open the serial connection
ser = serial.Serial(microbit_port, baud_rate, timeout=1)

# Give the Micro:bit some time to reset
time.sleep(2)

# Example: Send data to the Micro:bit
ser.write(b'Hello, Micro:bit!\n')

# Example: Read data from the Micro:bit
while True:
    if ser.in_waiting > 0:
        line = ser.readline().decode('utf-8').rstrip()
        print(line)
```

```python
def on_ir_button_released():
    global LiftedBUTTON, LAST, ReadCOMMAND
    LiftedBUTTON = makerbit.ir_button()
    LAST = -1
    ReadCOMMAND = -2
makerbit.on_ir_button(IrButton.ANY, IrButtonAction.RELEASED, on_ir_button_released)

def on_ir_button_pressed():
    global ReadCOMMAND
    ReadCOMMAND = makerbit.ir_button()
makerbit.on_ir_button(IrButton.ANY, IrButtonAction.PRESSED, on_ir_button_pressed)

def doSetParameters():
    global BTN_LIFTED, LiftedBUTTON, LIGHT, FWD, MOMENTUM, MySPEED, FSPEED, SPDMAX, TRIM, TTRIM, LEFT, RIGHT, TSPEED, LastCOUNT, LAST, COMMAND, TurnCOUNT
    BTN_LIFTED = -1
    LiftedBUTTON = -1
    LIGHT = False
    FWD = True
    MOMENTUM = -4
    MySPEED = 100
    FSPEED = 120
    SPDMAX = 250
    TRIM = 20
    TTRIM = 10
    LEFT = FSPEED
    RIGHT = FSPEED
    TSPEED = 100
    LastCOUNT = 0
    LAST = -1
    COMMAND = -2
    TurnCOUNT = 0
def Set_Button_Codes_Here():
    global ALL_STOP, SetLIGHT, FORWARD, INC_FWD_SPEED, POWER_BUTTON, INC_LEFT_TURN, INC_RIGHT_TURN, BACKWARDS, RESET_SPEED, INC_SPEED, REDUCE_SPEED, ROTATE_LEFT, ROTATE_RIGHT, MENU_BUTTON
    ALL_STOP = 160
    SetLIGHT = 64
    FORWARD = 128
    INC_FWD_SPEED = 48
    POWER_BUTTON = 0
    INC_LEFT_TURN = 32
    INC_RIGHT_TURN = 96
    BACKWARDS = 144
    RESET_SPEED = 88
    INC_SPEED = 48
    REDUCE_SPEED = 112
    ROTATE_LEFT = 16
    ROTATE_RIGHT = 80
    MENU_BUTTON = -1
MENU_BUTTON = 0
ROTATE_RIGHT = 0
ROTATE_LEFT = 0
REDUCE_SPEED = 0
INC_SPEED = 0
RESET_SPEED = 0
BACKWARDS = 0
INC_RIGHT_TURN = 0
INC_LEFT_TURN = 0
POWER_BUTTON = 0
INC_FWD_SPEED = 0
FORWARD = 0
SetLIGHT = 0
ALL_STOP = 0
TurnCOUNT = 0
COMMAND = 0
LastCOUNT = 0
TSPEED = 0
RIGHT = 0
LEFT = 0
TTRIM = 0
TRIM = 0
SPDMAX = 0
FSPEED = 0
MySPEED = 0
MOMENTUM = 0
LIGHT = False
BTN_LIFTED = 0
ReadCOMMAND = 0
LAST = 0
LiftedBUTTON = 0
FWD = False
FWD = True
makerbit.connect_ir_receiver(DigitalPin.P8, IrProtocol.NEC)
Set_Button_Codes_Here()
doSetParameters()
basic.show_icon(IconNames.HAPPY)
basic.pause(1000)
basic.clear_screen()
led.enable(True)

def on_forever():
    global BTN_LIFTED, LiftedBUTTON, COMMAND, FWD, FSPEED, TSPEED, LIGHT, LAST
    BTN_LIFTED = LiftedBUTTON
    LiftedBUTTON = -1
    if BTN_LIFTED < 0:
        pass
    elif BTN_LIFTED == 0:
        basic.show_icon(IconNames.TORTOISE)
        basic.clear_screen()
    elif BTN_LIFTED == RESET_SPEED:
        mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
        basic.show_icon(IconNames.HAPPY)
        basic.clear_screen()
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
    elif BTN_LIFTED == ALL_STOP:
        mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
        basic.show_icon(IconNames.CHESSBOARD)
        music.play_tone(131, music.beat(BeatFraction.HALF))
        basic.clear_screen()
    elif BTN_LIFTED == INC_LEFT_TURN:
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
        mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
    elif BTN_LIFTED == INC_RIGHT_TURN:
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
        mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
    elif BTN_LIFTED == BACKWARDS:
        mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
    elif BTN_LIFTED == FORWARD:
        mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
    elif BTN_LIFTED == REDUCE_SPEED:
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
    elif BTN_LIFTED == INC_SPEED:
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
    elif BTN_LIFTED == ROTATE_LEFT:
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
        mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
    elif BTN_LIFTED == ROTATE_RIGHT:
        mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
        mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
    elif BTN_LIFTED == SetLIGHT:
        pass
    else:
        basic.show_number(BTN_LIFTED)
    BTN_LIFTED = -1
    COMMAND = ReadCOMMAND
    if COMMAND == LAST:
        pass
    else:
        if COMMAND < 0:
            pass
        elif COMMAND == 0:
            pass
        elif COMMAND == RESET_SPEED:
            doSetParameters()
        elif COMMAND == ALL_STOP:
            mbit_Robot.car_ctrl(mbit_Robot.CarState.CAR_STOP)
        elif COMMAND == INC_LEFT_TURN:
            mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.GREEN)
            mbit_Robot.car_ctrl_speed(mbit_Robot.CarState.CAR_LEFT, TSPEED)
        elif COMMAND == INC_RIGHT_TURN:
            mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.BLUE)
            mbit_Robot.car_ctrl_speed(mbit_Robot.CarState.CAR_RIGHT, TSPEED)
        elif COMMAND == BACKWARDS:
            if LIGHT:
                mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.RED)
            else:
                pass
            FWD = False
            mbit_Robot.car_ctrl_speed(mbit_Robot.CarState.CAR_BACK, FSPEED)
        elif COMMAND == FORWARD:
            if LIGHT:
                mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.WHITE)
            else:
                pass
            FWD = True
            mbit_Robot.car_ctrl_speed(mbit_Robot.CarState.CAR_RUN, FSPEED)
        elif COMMAND == REDUCE_SPEED:
            basic.show_arrow(ArrowNames.SOUTH)
            if FSPEED > TRIM:
                FSPEED += 0 - TRIM
            else:
                FSPEED = 0
            basic.show_number(FSPEED)
            if TSPEED > TTRIM:
                TSPEED += 0 - TTRIM
            else:
                TSPEED = 20
            basic.clear_screen()
        elif COMMAND == INC_SPEED:
            basic.show_arrow(ArrowNames.NORTH)
            if FSPEED < SPDMAX:
                FSPEED += TRIM
            else:
                FSPEED = SPDMAX
            basic.show_number(FSPEED)
            if TSPEED < SPDMAX:
                TSPEED += TTRIM
            else:
                TSPEED = SPDMAX
            basic.clear_screen()
        elif COMMAND == ROTATE_LEFT:
            mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.CYAN)
            mbit_Robot.car_ctrl_speed(mbit_Robot.CarState.CAR_SPINLEFT, TSPEED)
        elif COMMAND == ROTATE_RIGHT:
            mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.PINKISH)
            mbit_Robot.car_ctrl_speed(mbit_Robot.CarState.CAR_SPINRIGHT, TSPEED)
        elif COMMAND == SetLIGHT:
            if LIGHT:
                mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.OFF)
                LIGHT = False
            else:
                mbit_Robot.RGB_Car_Big2(mbit_Robot.enColor.WHITE)
                LIGHT = True
        else:
            basic.show_number(COMMAND)
            basic.pause(100)
        LAST = COMMAND
basic.forever(on_forever)
```
